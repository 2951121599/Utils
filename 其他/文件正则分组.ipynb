```python
# PEPMASS (mz)
import os
import re

for root,dirs,files in os.walk(r"./mgf"):
    # print("root,dirs,files----------", root,dirs,files)
    
    for file in files:
        file_object = open(os.path.join(root,file), 'r', encoding='UTF-8')
        data = []

        params = []
        pep_idx_list = []
        try:
            for idx,line in enumerate(file_object):
                g = re.match("(^BEGIN IONS)", line)
                p = re.match("(^PEPMASS)", line)
                e = re.match("(^END IONS)", line)

                if g:
                    start = idx + 1
                    params.append(start)
                if p:
                    p_idx = idx + 1
                if e:
                    end = idx + 1
                    params.append(end)

                if line[0:7] == 'PEPMASS':
                    PEPMASS = line.split(' ')[0].split('=')[1]
                    PEPMASS = int(float(PEPMASS))
                    # print("PEPMASS----------", PEPMASS)
                    if 380 < PEPMASS < 520:
                        pep_idx_list.append(idx + 1)
            print(params)
        finally:
            file_object.close()
        ww = []
        for x in pep_idx_list:
            if (x -3) in params:
                ww.append(params.index(x-3))
        # print("ww----------", ww)
        with open(os.path.join(root,file),'r') as f:
            line=f.readlines()
        total = len(ww)
        for index,w in enumerate(ww):
            print("index,total--------------------", index+1,total)

            for i,rows in enumerate(line):
                if i in range(params[w]-1,params[w+1]) :  # 
                    data.append(rows)
            # print(data)
        # 
        with open(os.path.join("./new_mgf/35V_Pos_only_mz.txt"),"w",) as f:
          for i in range(len(data)):
              f.writelines(data[i])
        print("OK!")
```
